```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
 soup = BeautifulSoup(webpage , 'lxml')  # lxml is html traversal used for traversing html data it is mandatory to give it to BeaitifulSoup .
```

```python
# soup.prettify()    

# soup.prettify() method formats the html documnet and content we can see that by printing it like this print(soup.prettify())
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())     # here strip() is used for filling the gaps and formats the data
```

```python
company = soup.find_all('div',class_ = 'companyCardWrapper')
```
