```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
#extracting the numeric data
df['number_numerical'] = pd.to_numeric(df["number"] , errors = 'coerce' , downcast = 'integer')

#extracting the categorical data from the same column
df['number_categorical'] = np.where(df['number_numerical'].isnull() , df['number'] , np.nan)  #- np.where(condition, value_if_true, value_if_false)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')  # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] #captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind = 'bar')
```

```python
#extract the last bit of ticket as number

df['ticket_num'] = df['Ticket'].apply(lambda s : s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                errors = 'coerce',
                                downcast = 'integer')

# extract the first part of the ticket as ctegory

df['ticket_cat'] = df['Ticket'].apply(lambda s : s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit() , np.nan,df['ticket_cat'])

df.head()
```
